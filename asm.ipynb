```python
def Nhap_xuat_nhan_vien():
    print("=== Nhập và xuất danh sách nhân viên ===")

def Xoa_cap_nhat_thong_tin():
    print("=== Xóa hoặc cập nhật thông tin nhân viên ===")

def Tim_kiem_theo_luong():
    print("=== Tìm kiếm nhân viên theo lương ===")

def Sap_xep_ho_ten_va_thu_nhap():
    print("=== Sắp xếp nhân viên theo họ tên và thu nhập ===")

def top5_thu_nhap_cao():
    print("=== Xuất 5 nhân viên có thu nhập cao nhất công ty ===")

def Thoat():
    print("=== Thoát chương trình ===")

Menu = {
    "1": "Nhập và xuất danh sách nhân viên",
    "2": "Xóa hoặc cập nhật thông tin nhân viên",
    "3": "Tìm kiếm nhân viên theo lương",
    "4": "Sắp xếp nhân viên theo họ tên và thu nhập",
    "5": "Xuất 5 nhân viên có thu nhập cao nhất công ty",
    "6": "Thoát"
}

while True:
    print("\n" + "=" * 50)
    print("HỆ THỐNG QUẢN LÝ NHÂN VIÊN".center(50))
    print("=" * 50)

    for key, value in Menu.items():
        print(f"{key}. {value}")

    lua_chon = input("\nNhập lựa chọn của bạn: ")

    if lua_chon == "1":
        Nhap_xuat_nhan_vien()
    elif lua_chon == "2":
        Xoa_cap_nhat_thong_tin()
    elif lua_chon == "3":
        Tim_kiem_theo_luong()
    elif lua_chon == "4":
        Sap_xep_ho_ten_va_thu_nhap()
    elif lua_chon == "5":
        top5_thu_nhap_cao()
    elif lua_chon == "6":
        Thoat()
        break
    else:
        print(" Lựa chọn không hợp lệ vui lòng chọn lại")
```
